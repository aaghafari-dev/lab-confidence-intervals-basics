## Challenge 1
We want to estimate the average size of the men of a country with a confidence level of 80%. Assuming that the standard deviation of the sizes in the population is 4, get the confidence interval with a sample of men selected randomly, whose heights are:

167 167 168 168 168 169 171 172 173 175 175 175 177 182 195

```python
# import lib
import numpy as np
import scipy.stats as stats
import math
```

```python
heights = [167, 167, 168, 168, 168, 169, 171, 172, 173, 175, 175, 175, 177, 182, 195]
n = len(heights)
sample_mean = np.mean(heights)
sigma_pop = 4  # given value
confidence = 0.80  # given value

# z-critical for 80% confiddence
z = stats.norm.ppf(1 - (1 - confidence) / 2)  
margin = z * sigma_pop / np.sqrt(n)

ci_lower = sample_mean - margin
ci_upper = sample_mean + margin

print(f"Sample mean: {sample_mean:.2f}")
print(f"z-critical for 80% confidence: {z:.4f}")
print(f"80% Confidence interval for mean height: ({ci_lower:.2f}, {ci_upper:.2f})")
```

## Challenge 2 
In a sample of 105 shops selected randomly from an area, we note that 27 of them have had losses in this month. Get an confidence interval for the proportion of businesses in the area with losses, to a confidence level 
* of 80% and confidence level.
* of 90%.


```python
n = 105
losses = 27
p_hat = losses / n
q_hat = 1 - p_hat
se = np.sqrt(p_hat * q_hat / n)

# For 80% confidence
z80 = stats.norm.ppf(0.9)  # 80% CI leaves 10% in each tail
margin80 = z80 * se
ci80_lower = p_hat - margin80
ci80_upper = p_hat + margin80

# For 90% confidence
z90 = stats.norm.ppf(0.95)  # 90% CI leaves 5% in each tail
margin90 = z90 * se
ci90_lower = p_hat - margin90
ci90_upper = p_hat + margin90

print(f"Sample proportion: {p_hat:.4f}")
print(f"z-critical for 80% confidence: {z80:.4f}")
print(f"80% CI for proportion: ({ci80_lower:.4f}, {ci80_upper:.4f})")
print(f"\nz-critical for 90% confidence: {z90:.4f}")
print(f"90% CI for proportion: ({ci90_lower:.4f}, {ci90_upper:.4f})")
```

```python
heights = [167, 167, 168, 168, 168, 169, 171, 172, 173, 175, 175, 175, 177, 182, 195]
n = len(heights)
sample_var = np.var(heights, ddof=1)  
df = n - 1
confidence = 0.90
alpha = 1 - confidence

# Chi-squared critical values
chi2_lower = stats.chi2.ppf(1 - alpha/2, df)  # for lower bound denominator
chi2_upper = stats.chi2.ppf(alpha/2, df)      # for upper bound denominator

ci_var_lower = (n - 1) * sample_var / chi2_lower
ci_var_upper = (n - 1) * sample_var / chi2_upper

print(f"Sample variance: {sample_var:.4f}")
print(f"Chi-squared lower critical (df={df}, {alpha/2:.2f} upper tail): {chi2_lower:.4f}")
print(f"Chi-squared upper critical (df={df}, {alpha/2:.2f} lower tail): {chi2_upper:.4f}")
print(f"90% Confidence interval for population variance: ({ci_var_lower:.4f}, {ci_var_upper:.4f})")
```

## Challenge 4 - To practice more
The sulfuric acid content of 7 similar containers is 9.8, 10.2, 10.4, 9.8, 10.0, 10.2 and 9.6 liters. Calculate a 95% confidence interval for the average content of all containers assuming an approximately normal distribution.




```python
acid = [9.8, 10.2, 10.4, 9.8, 10.0, 10.2, 9.6]
n = len(acid)
sample_mean = np.mean(acid)
sample_std = np.std(acid, ddof=1)
df = n - 1
confidence = 0.95

# t-critical
t = stats.t.ppf(1 - (1 - confidence) / 2, df)
margin = t * sample_std / np.sqrt(n)

ci_lower = sample_mean - margin
ci_upper = sample_mean + margin

print(f"Sample mean: {sample_mean:.2f}")
print(f"t-critical (df={df}) for 95% confidence: {t:.4f}")
print(f"95% Confidence interval for mean content: ({ci_lower:.2f}, {ci_upper:.2f})")
```
